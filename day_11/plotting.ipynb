```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data set
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python

titanic['survived'].value_counts()
```

## Bar plot

```python
titanic['survived'].value_counts().plot(kind='bar')
```

```python
# plot using sns
sns.countplot(x='survived', data=titanic)
```

```python
sns.countplot(x='survived', hue='sex', data=titanic)
```

```python
sns.barplot(data=titanic, x='sex', y='fare', hue='pclass')
```

## Scatter Plot

```python
plt.scatter(data=titanic, x='age', y='fare')
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='sex')
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='pclass')
```

## Pair Plot

```python
sns.pairplot(titanic)
```

## Hist plot

```python
titanic.hist(bins=10, figsize=(10, 10))
```
