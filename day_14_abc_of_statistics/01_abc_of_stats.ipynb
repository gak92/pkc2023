# Statistics

- Application of Statistics
- Descriptive Statistics (Presenting, Organizing, and summarizing)
- Inferential Statistics (Drawing conclusions about a population based on data observed in a sample)

- `Data`: Facts (especially numerical), collected together for reference or information
- `Information`: Knowledge communicated concerning some particular facts

> Q: What is statistics?\
A: The science of uncertainty.\
   Methods for collecting, analyzing, interpreting, and presenting empirical data.

- Two fundamental ideas:
1. Uncertainty
2. Variation

## 3 Stages of Statistics
1. Design / Planning
2. Descriptive
3. Inferential
